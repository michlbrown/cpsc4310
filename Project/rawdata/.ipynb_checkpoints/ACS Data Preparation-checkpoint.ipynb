#### Removing extra colums
For now, we will be removing the columns containing error values and the percentage values, we will keep this in raw data for use later.

Valid columns that contain only the estimates end with a singular E 
e.g. DP02_0001E, DP05_0089E)

We are using a regex to filter only the estimate columns, while keeping the GEO_ID and NAME values

```python
acs_data = acs_data.filter(regex="^(GEO_ID)|(NAME)|(DP\d\d_\d\d\d\d[E])$")
```

```python
acs_data = pd.read_csv("trimmed_acs.csv", dtype=str)

raw_labels = acs_data.loc[29567, :]
raw_labels = raw_labels[2 : ]

# Example of raw label
print(raw_labels[0])
```

```python
import re

# Remove the redundant Estimate and !! values
regex1 = "^Estimate!!"
raw_labels = [re.sub(regex1, "", label) for label in raw_labels]
raw_labels = [re.sub("!!", ", ", label) for label in raw_labels]

# Map coded lable to readable lable
labels_dict = dict()

for col, label in zip(acs_data.columns[2:], raw_labels):     # ignoring the first 2 columns as they are already readable
    labels_dict[col] = label
    
labels = pd.Series(labels_dict)

labels.to_csv("acs_lables.csv")
```
