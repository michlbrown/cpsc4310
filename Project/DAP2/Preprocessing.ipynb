# Data Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("../rawdata/acs.csv", index_col=0)
```

## Separate identifiers (GEO_ID and NAME) into their own table

```python
data_names = data[["GEO_ID", "NAME"]]
data = data.drop(columns=["GEO_ID", "NAME"])
```

## Remove remaining non-numeric data

```python
data.dtypes
```

```python
data = data.apply(pd.to_numeric, errors="coerce")
```

## Remove NaN values

```python
data = data.dropna()
```

```python
data
```

## Descriptive Stats

```python
data["total_pop"].describe()
```

```python
# show a histogram and boxplot for every column
for col in data.columns:
    print(f"{col}: ")
    plt.subplot(1,2,1)
    data[col].plot.hist()
    plt.xlabel(col)

    plt.subplot(1,2,2)
    data[col].plot.box()
    plt.show()
    
    print(data[col].describe())
    print()
```

```python
# see what percent of the data is above the national poverty level
len([x for x in data["p_families_poverty"] if x > 11.8])/len(data["p_families_poverty"])
```

```python
# split the data into 2 bins based on poverty rate
from sklearn.preprocessing import Binarizer
X = data["p_families_poverty"].to_numpy()
est = Binarizer(threshold=11.8)
X= X.reshape(-1,1)
Xt = est.fit_transform(X)
data.insert(0, column="poverty", value=Xt)
```

```python
len([x for x in data["poverty"] if x == 0])/len(data["poverty"])
```

# Analysis

```python
for col in data:
    data.plot.scatter(x=col, y="p_families_poverty", s=1)
```

```python
for col in data:
    print(data[col].corr(data["poverty"], method="pearson"), col)
```

```python
# print out the medians for both classes
poverty_above_stats = data.loc[data["poverty"] == 1].describe()
poverty_below_stats = data.loc[data["poverty"] == 0].describe()
```

```python
# print the medians of both classes
for col in poverty_above_stats:
    print("median of", col)
    print("1: ", poverty_above_stats[col]["50%"])
    print("0: ", poverty_below_stats[col]["50%"])
    print()
```

```python
# visualize the differences between the classes
#poverty = data.loc[data["poverty"] == 1]
#no_poverty = data.loc[data["poverty"] == 0]
for col in data:
    data.boxplot(column=col, by="poverty")
    plt.show()
```
